```python
import os, shutil
import io
from io import StringIO
import re
import sys
from contextlib import redirect_stdout
from keras import models
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as sp
from scipy.stats import kde
import importlib as im
import cnn_utilities as cn
from sklearn import metrics
import csv
```

```python
# load trained model and normalization values
my_model = models.load_model("saved_models/MTBD_128batch_20epoch.hdf5")

my_root_model = models.load_model("saved_models/MTBD_128batch_12epoch_root_location.hdf5", compile = False)
my_root_model.compile(loss = 'categorical_crossentropy', optimizer = 'rmsprop', 
                      metrics = ['acc', tf.keras.metrics.AUC(), tf.keras.metrics.CategoricalCrossentropy()])

mean_sd = pd.read_csv("saved_models/train_mtbd_128batch_20epoch_normalization_label_mean_sd.csv", 
                     index_col = 0).to_numpy()
```

```python
train_means = mean_sd[0,:]
train_sd = mean_sd[1,:]
train_aux_priors_means = train_means[7:,]
train_aux_priors_sd = train_sd[7:,]

num_locs = 5
max_tips = 502
```

```python
###################
# true model test #
###################

mtbd_phylocomp_data = pd.read_csv("data_files/mtbd.cblv.csv", 
                   header =None, error_bad_lines = False, index_col = 0).to_numpy()

mtbd_phylocomp_labels = pd.read_csv("data_files/mtbd_labels.txt",
                    header = None, error_bad_lines = False).to_numpy()

mtbd_phylo_post = pd.read_table("data_files/mtbd_rates.postmeans", 
                   header = None, index_col = 0, error_bad_lines = False).to_numpy()

mtbd_phylo_post_root_freq = pd.read_table("data_files/mtbd_root_probs.tsv", 
                   header = None, index_col = 0, error_bad_lines = False).to_numpy()


```

```python
mtbd_phylocomp_labels.shape
```

```python
#########################
# prepare input tensors #
#########################

# normalize with training means and sds
mtbd_phylocomp_subsample_prop = mtbd_phylocomp_data[:,(max_tips-1) * 7]
mtbd_phylocomp_mu = mtbd_phylocomp_data[:,(max_tips-3) * 7]
mtbd_phylocomp_num_tips = cn.get_num_tips(mtbd_phylocomp_data)
mtbd_phylocomp_tmrca = mtbd_phylocomp_labels[:,12]
mtbd_phylocomp_mean_bl = mtbd_phylocomp_labels[:,13]

aux_mtbd_phylocomp = np.vstack((mtbd_phylocomp_mu, mtbd_phylocomp_subsample_prop, mtbd_phylocomp_num_tips,
                              mtbd_phylocomp_tmrca, mtbd_phylocomp_mean_bl)).transpose()
norm_aux_mtbd_phylocomp = cn.normalize(aux_mtbd_phylocomp, (train_aux_priors_means, train_aux_priors_sd))

mtbd_phylocomp_treeLocation_tensor, mtbd_phylocomp_prior_tensor = cn.create_data_tensors(mtbd_phylocomp_data,
                                                                          mu = norm_aux_mtbd_phylocomp[:,0],
                                                                          subsample_prop = norm_aux_mtbd_phylocomp[:,1],
                                                                            num_tips = norm_aux_mtbd_phylocomp[:,2],
                                                                            tmrca = norm_aux_mtbd_phylocomp[:,3],
                                                                            mean_bl = norm_aux_mtbd_phylocomp[:,4],
                                                                            num_locs = num_locs,
                                                                            max_tips = max_tips,
                                                                            cblv_contains_mu_rho = True)
```

```python
# predict phylodynamics
mtbd_phylocomp_normalized_preds = my_model.predict([mtbd_phylocomp_treeLocation_tensor, 
                                         mtbd_phylocomp_prior_tensor])

# denormalize preds use label means and sd from training data normalization
cnn_mtbd_phylocomp_preds = cn.denormalize(mtbd_phylocomp_normalized_preds, train_means[0:7], train_sd[0:7])
cnn_mtbd_phylocomp_preds = np.exp(cnn_mtbd_phylocomp_preds)

# measure error
mtbd_phylocomp_abserror = np.abs(np.subtract(cnn_mtbd_phylocomp_preds, mtbd_phylocomp_labels[:,5:12]))
mtbd_phylocomp_mape = np.multiply(100, np.abs(np.divide(mtbd_phylocomp_abserror, mtbd_phylocomp_labels[:,5:12])))

# predict root location
mtbd_phylocomp_root_preds = my_root_model.predict([mtbd_phylocomp_treeLocation_tensor, 
                                                          mtbd_phylocomp_prior_tensor])
```

```python
# save predictions to file for analysis in R 
# CNN predictions
df_mtbd_phylocomp_cnn_output = pd.DataFrame(np.concatenate((cnn_mtbd_phylocomp_preds, mtbd_phylocomp_root_preds), axis = 1), 
                                           columns = ["R0_0", "R0_1", "R0_2", "R0_3", "R0_4",
                                                      "sample_rate", "migration_rate",
                                                      "Loc1", "Loc2", "Loc3", "Loc4", "Loc5"])
df_mtbd_phylocomp_cnn_output.to_csv('output/mtbd_phylocomp_cnn_preds.tsv', sep = "\t", index = False)


# true values
df_mtbd_phylocomp_true_output = pd.DataFrame(np.concatenate((mtbd_phylo_post, mtbd_phylo_post_root_freq), axis = 1),
                                        columns = ["R0_0", "R0_1", "R0_2", "R0_3", "R0_4",
                                                   "sample_rate", "migration_rate",
                                                      "Loc1", "Loc2", "Loc3", "Loc4", "Loc5"])
df_mtbd_phylocomp_true_output.to_csv('output/mtbd_phylocomp_phylo_means.tsv', sep = '\t', index = False)


# Phylogeographical predictions
df_mtbd_phylocomp_phylo_output = pd.DataFrame(np.concatenate((mtbd_phylocomp_labels[:,5:12], mtbd_phylocomp_labels[:,:5]), axis = 1),
                                        columns = ["R0_0", "R0_1", "R0_2", "R0_3", "R0_4",
                                                   "sample_rate", "migration_rate",
                                                      "Loc1", "Loc2", "Loc3", "Loc4", "Loc5"])
df_mtbd_phylocomp_phylo_output.to_csv('output/mtbd_phylocomp_labels.tsv', sep = '\t', index = False)

```
