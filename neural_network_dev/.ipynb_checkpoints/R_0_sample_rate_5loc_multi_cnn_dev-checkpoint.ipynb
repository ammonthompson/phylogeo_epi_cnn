```python
import os, shutil
from keras import models
from keras import layers
from keras import losses
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as sp
```

```python
full_data = pd.read_csv("data_files/set_0to25.cblv.csv", 
                   header =None, error_bad_lines = False).to_numpy()

full_labels = pd.read_csv("data_files/loc_R0_delta_migration_set_0to25.csv",
                    header = None, error_bad_lines = False).to_numpy()
```

```python
# randomize data order
randomized_idx = np.random.permutation(full_data.shape[0])
full_data = full_data[randomized_idx,:]
full_labels = full_labels[randomized_idx,:]
```

```python
# # R0 in a range
# r0_idx = np.where((full_labels[:,5] > 5))[0]
# full_data = full_data[r0_idx,:]
# full_labels = full_labels[r0_idx,:]
```

```python
full_data.shape
```

```python
# normalize labels
label_means = full_labels.mean(axis=0)
label_sd = full_labels.std(axis=0)

#full_labels -= label_means
#full_labels /= label_sd

norm_full_labels = (full_labels - label_means) / label_sd
```

```python
# setting up the data
num_locs = 5
max_tips = 502
num_sample = full_data.shape[0]


num_test = 2000
num_validation = 2000

data = full_data[0:num_sample,:]
```

```python
# make training tensors
full_data_tensor = data.reshape((num_sample, max_tips, num_locs+2), order = 'C')


### divide full_data_tensor into tree, locations, and prior tensors

#tree and location tensor
full_treeLocation_tensor = full_data_tensor[:,:max_tips-3,:]

# prior tensor
    # transform subsample proportion and mu to fix sensitivity
logit_subprop = np.log(full_data_tensor[:,max_tips-1,0] / (1 - full_data_tensor[:,max_tips-1,0] + 0.0001))
shifted_log_subprop = np.min(full_data_tensor[:,max_tips-1,0]) + np.log(full_data_tensor[:,max_tips-1,0])
linear_subprop = 2.5 * full_data_tensor[:,max_tips-1,0]
    # proportion sampled and recover(death) rate
subprop = np.repeat(linear_subprop, 2)
subprop = subprop.reshape((num_sample, 1, 2))
mu = np.repeat(50 * full_data_tensor[:,max_tips-3,0] , 2)
mu = mu.reshape((num_sample , 1, 2))
    # tree size
num_tips = []
for i in range(full_data_tensor.shape[0]):
    num_tips.append(1+len(np.where(full_data_tensor[i,:,0] > 0)[0]))
num_tips = np.asarray(num_tips)
norm_nt = (num_tips - np.mean(num_tips))/np.std(num_tips)
norm_num_tips = np.repeat(norm_nt, 2)
norm_num_tips = norm_num_tips.reshape((num_sample, 1, 2))

    # put them together
full_prior_tensor = np.concatenate((mu, subprop, norm_num_tips), axis = 1)

# location tensor
full_location_tensor = full_data_tensor[:,:max_tips-3,2:8]

# tree tensor
full_tree_tensor = full_data_tensor[:,:max_tips-3,0:2]

```

```python
# training tensors
train_treeLocation_tensor = full_treeLocation_tensor[num_test + num_validation:,:,:]
train_prior_tensor = full_prior_tensor[num_test + num_validation:,:,:]
train_location_tensor = full_location_tensor[num_test + num_validation:,:,:]
train_tree_tensor = full_tree_tensor[num_test + num_validation:,:,:]

# validation tensors
validation_treeLocation_tensor = full_treeLocation_tensor[num_test:num_test + num_validation,:,:]
validation_prior_tensor = full_prior_tensor[num_test:num_test + num_validation,:,:]
validation_location_tensor = full_location_tensor[num_test:num_test + num_validation,:,:]
validation_tree_tensor = full_tree_tensor[num_test:num_test + num_validation,:,:]

# testing tensors
test_treeLocation_tensor = full_treeLocation_tensor[:num_test,:,:]
test_prior_tensor = full_prior_tensor[:num_test,:,:]
test_location_tensor = full_location_tensor[:num_test,:,:]
test_tree_tensor = full_tree_tensor[:num_test,:,:]
```

```python
print(train_treeLocation_tensor.shape, train_prior_tensor.shape, train_location_tensor.shape, train_tree_tensor.shape)
print(test_treeLocation_tensor.shape, test_prior_tensor.shape, test_location_tensor.shape, test_tree_tensor.shape)
print(validation_treeLocation_tensor.shape, validation_prior_tensor.shape, validation_location_tensor.shape,
      validation_tree_tensor.shape)
```

```python
# training and validation labels
# indices for R0, sample_rate, and migration rate
# are 5:8. 9 is the subsampling proportion
##################################

labels = norm_full_labels[0:num_sample,5:8]
train_labels = labels[num_test+num_validation:,]
test_labels = labels[:num_test,]
validation_labels = labels[num_test:num_test+num_validation,]
```

```python
print(test_labels.shape)
print(train_labels.shape)
print(validation_labels.shape)
```

```python
# multi input cnn
from keras import Input
from keras import Model

num_train_minus_val_samples = train_labels.shape[0]

# tree + geological deep conv1D
input_treeLocation_tensor = Input(shape = train_treeLocation_tensor.shape[1:3])
w1 = layers.Conv1D(84, 3, strides = 1, activation = 'relu', padding = 'same')(input_treeLocation_tensor)
w = layers.Conv1D(84, 3, dilation_rate = 1, activation = 'relu', padding = 'same')(w1)
w = layers.Conv1D(128, 4, dilation_rate = 1, activation = 'relu', padding = 'same')(w)
# w = layers.MaxPooling1D(pool_size = 3, stride = 2)(w)
# w = layers.BatchNormalization()(w)
w = layers.SpatialDropout1D(rate = 0.2)(w)
w = layers.GlobalAveragePooling1D(name = 'w')(w)

# tree + geolocation dilated
w_dilated = layers.Conv1D(64, 3, dilation_rate = 2, activation = 'relu', padding = "same")(w1)
w_dilated = layers.Conv1D(64, 4, dilation_rate = 4, activation = 'relu', padding = "same")(w_dilated)
w_dilated = layers.Conv1D(128, 4, dilation_rate = 4, activation = 'relu', padding = "same")(w_dilated)
w_dilated = layers.SpatialDropout1D(rate = 0.2)(w_dilated)
w_dilated = layers.GlobalAveragePooling1D(name = 'w_dilated')(w_dilated)

# # tree only
input_tree_tensor = Input(shape = train_tree_tensor.shape[1:3])
tree = layers.Conv1D(64, 3, strides = 1, activation = 'relu', padding = 'same')(input_tree_tensor)
tree = layers.Conv1D(128, 4, strides = 2, activation = 'relu', padding = 'same')(input_tree_tensor)
tree = layers.SpatialDropout1D(rate = 0.2)(tree)
tree = layers.Conv1D(1, 1, activation = 'elu', padding = 'same')(tree)
tree = layers.Flatten()(tree)

# geolocation only
input_location_tensor = Input(shape = train_location_tensor.shape[1:3])
geo = layers.Conv1D(32, 7, strides = 5, activation = 'elu', padding = 'same')(input_location_tensor)
geo = layers.Conv1D(1,1, activation ='elu', padding = 'same')(geo)
geo = layers.Flatten()(geo)

# # concatenate tree and geo above -> deep conv1D network
# tree_geo_concatenated = layers.Concatenate(axis = 2)([tree, geo])
# tree_geo_concatenated = layers.Conv1D(32, 3, strides = 1, activation = 'relu', padding = 'same')(tree_geo_concatenated)
# tree_geo_concatenated = layers.Conv1D(32, 3, dilation_rate = 2, activation = 'relu', padding = 'same')(tree_geo_concatenated)
# tree_geo_concatenated = layers.Conv1D(64, 3, dilation_rate = 4, activation = 'relu', padding = 'same')(tree_geo_concatenated)
# tree_geo_concatenated = layers.Conv1D(1, 1, activation = 'relu', padding = 'same', name = 'treegeo_reduced')(tree_geo_concatenated)
# tree_geo_concatenated = layers.Flatten(name = 'tree_geo_flattened')(tree_geo_concatenated)

# prior known parameters and data statistics
input_priors_tensor = Input(shape = train_prior_tensor.shape[1:3])
priors = layers.Flatten()(input_priors_tensor)
priors = layers.Dense(64, activation = 'relu', kernel_initializer = 'VarianceScaling', name = 'prior')(priors)

# concatenate all above -> deep fully connected network
concatenated_wxyz = layers.Concatenate(axis = 1, name = 'all_concatenated')([w, w_dilated, tree, geo, priors])
#wxyz = layers.Masking(mask_value = 0.0)(concatenated_wxyz)
wxyz = layers.Dense(200, activation = 'relu', kernel_initializer = 'VarianceScaling')(concatenated_wxyz)
wxyz = layers.Dense(100, activation = 'relu', kernel_initializer = 'VarianceScaling')(wxyz)
wxyz = layers.Dense(50, activation = 'relu', kernel_initializer = 'VarianceScaling')(wxyz)
wxyz = layers.Dense(30, activation = 'relu', kernel_initializer = 'VarianceScaling')(wxyz)
#wxyz = layers.BatchNormalization()(wxyz)
#wxyz = layers.Dropout(rate = 0.2)(wxyz)

output_R0_sampleRate_migrationRate = layers.Dense(3, activation = 'linear')(wxyz)

```

```python
mymodel = Model(inputs = [input_treeLocation_tensor, input_tree_tensor, input_priors_tensor, input_location_tensor], 
                outputs = output_R0_sampleRate_migrationRate)
mymodel.compile(optimizer = 'rmsprop', 
                loss = 'mse', 
                metrics = ['mae', 'acc', 'mape'])
history = mymodel.fit([train_treeLocation_tensor, train_tree_tensor, train_prior_tensor, train_location_tensor], 
                      train_labels,
                    epochs = 8, batch_size = 32, 
                    validation_data = ([validation_treeLocation_tensor, validation_tree_tensor, validation_prior_tensor, 
                                        validation_location_tensor], 
                                       validation_labels))

# get baseline mape distribution to compare with prior distro of mapes from likelihood model
# train_nodat = np.zeros(train_data_tensor_minus_val.shape)
# history = mymodel.fit(train_nodat, train_labels_minus_val,
#                      epochs = 15, batch_size = 64, 
#                      validation_data = (train_data_tensor_val, train_labels_val))
```

```python
#########################################
# compute errors on misspecified trees ##
#########################################
misspec_delta_data = pd.read_csv("data_files/misspec_delta.cblv.csv", 
                   header =None, error_bad_lines = False).to_numpy()
```

```python
#########################
# prepare input tensors #
#########################
num_misspec = misspec_delta_data.shape[0]
misspec_test_tensor = misspec_delta_data.reshape((misspec_delta_data.shape[0], max_tips, num_locs+2), order = 'C')

miss_num_sample = misspec_test_tensor.shape[0]

#tree and location tensor
misspec_treeLocation_tensor = misspec_test_tensor[:,:max_tips-3,:]

#prior tensor
    # tree size
miss_num_tips = []
for i in range(misspec_test_tensor.shape[0]):
    miss_num_tips.append(1+len(np.where(misspec_test_tensor[i,:,0] > 0)[0]))
miss_num_tips = np.asarray(miss_num_tips)
miss_norm_nt = (miss_num_tips - np.mean(miss_num_tips))/np.std(miss_num_tips)
miss_norm_num_tips = np.repeat(miss_norm_nt, 2)
miss_norm_num_tips = miss_norm_num_tips.reshape((miss_num_sample, 1, 2))

miss_logit_subprop = np.log(misspec_test_tensor[:,max_tips-1,0] / (1 - misspec_test_tensor[:,max_tips-1,0] + 0.0001))
miss_shifted_log_subprop = np.min(misspec_test_tensor[:,max_tips-1,0]) + np.log(misspec_test_tensor[:,max_tips-1,0])
miss_linear_subprop = 2.5 * misspec_test_tensor[:,max_tips-1,0]

miss_subprop = np.repeat(miss_linear_subprop , 2)
miss_subprop = miss_subprop.reshape((miss_num_sample, 1, 2))                          

miss_mu = np.repeat(50 * misspec_test_tensor[:,max_tips-3,0] , 2)
miss_mu =miss_mu.reshape(miss_num_sample , 1, 2)

misspec_prior_tensor = np.concatenate((miss_mu, miss_subprop, miss_norm_num_tips), axis = 1)

# location tensor
misspec_location_tensor = misspec_test_tensor[:,:max_tips-3,2:8]

# tree tensor
misspec_tree_tensor = misspec_test_tensor[:,:max_tips-3,0:2]

# load labels data
misspec_labels = pd.read_csv("data_files/R0_delta_migration_misspec_delta.csv",
                    header = None, error_bad_lines = False, index_col = 0).to_numpy()

# normalize labels
miss_label_means = misspec_labels.mean(axis=0)
miss_label_sd = misspec_labels.std(axis=0)
misspec_norm_labels = np.subtract(misspec_labels, miss_label_means)
misspec_norm_labels = np.divide(misspec_norm_labels, miss_label_sd)
```

```python
# predict
miss_normalized_preds = mymodel.predict([misspec_treeLocation_tensor, misspec_tree_tensor, 
                                         misspec_prior_tensor, misspec_location_tensor])

# denormalize
miss_denormalized_test_labels = np.array([ np.add(np.multiply(misspec_norm_labels[:,i], miss_label_sd[i]), 
                                                  miss_label_means[i]) for i in range(5,8)]).transpose()
miss_denormalized_preds = np.array([ np.add(np.multiply(miss_normalized_preds[:,i-5], miss_label_sd[i]), 
                                            miss_label_means[i]) for i in range(5,8)]).transpose()

# measure error
miss_abserror = np.abs(np.subtract(miss_denormalized_preds, miss_denormalized_test_labels))
miss_mape = np.multiply(100, np.abs(np.divide(miss_abserror, miss_denormalized_test_labels)))

```
